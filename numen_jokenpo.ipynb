# Descrição
### jokenpô

pedra
papel
tesoura

regras: 
- pedra vence tesoura
- tesoura vence papel
- papel vence pedra

1 - input - usuário seleciona umas das 3 opções  
2 - computador gera uma opção aleatória   
3 - programa compara seleções com base na regra do jogo  
4 - programa informa o resultado, inclusive se der empate  
5 - permitir que o usuário jogue várias vezes até que deseje sair  
6 - construir placar que acompanha o jogador durante a sessão ativa  
7 - adicionar um modo de jogo onde o jogador precisa vencer uma melhor de 3, uma melhor de 5, etc (e ser declarado o vencedor dessa sequencia)  
8 - estatísticas do final do jogo (número total de partidas jogadas, percentual de vitórias, derrotas e empates)

# Código

```python
import random

dicionario_melhor_visualizacao = {'1': 'Pedra', '2': 'Papel', '3': 'Tesoura'}


def menu_principal():
    
    print("\n")
    print('----- Bem-vindo ao Jo-Ken-Pô! -----')
    print('--------- MENU PRINCIPAL ----------')
    print('-------- 1 - Jogo Único ----------')
    print('-------- 2 - Vários Jogos --------')
    print('-------- 3 - Melhor de 3 ----------')
    print('-------- 4 - Melhor de 5 ----------')
    print('-------- 5 - Melhor de 7 ----------')
    print('-------- 6 - Sair ---------------')
    print("\n")
    

def menu_jogo(jogo_multiplo):
    if jogo_multiplo:
        print('-------- Faça sua jogada ---------- ')
        print('----------- 1 - Pedra ------------- ')
        print('----------- 2 - Papel ------------- ')
        print('----------- 3 - Tesoura ----------- ')
        print('----------- 4 - Sair -------------- ')
        print("\n")
    else:
        print('-------- Faça sua jogada ---------- ')
        print('----------- 1 - Pedra ------------- ')
        print('----------- 2 - Papel ------------- ')
        print('----------- 3 - Tesoura ----------- ')
        print("\n")
    
  
def jogadas(jogo_multiplo):
    while True:
        escolha_jogador = input('Escolha: ')
        print("\n")
        
        if jogo_multiplo: 
            
            if escolha_jogador == '4':
                return '4', None
                
            elif escolha_jogador not in dicionario_melhor_visualizacao:
                print("Escolha inválida. Tente novamente.")
                print("\n")
                menu_jogo(True)              
                
            else:
                escolha_jogador = dicionario_melhor_visualizacao[escolha_jogador]
                print(f"Você escolheu: {escolha_jogador}")            

                escolha_computador = random.choice(list(dicionario_melhor_visualizacao.values()))  
                print(f"O computador escolheu: {escolha_computador}")
                print("\n")
                return escolha_jogador, escolha_computador   
            
        else:
            if escolha_jogador not in dicionario_melhor_visualizacao:
                print("Escolha inválida. Tente novamente.")
                print("\n")
                menu_jogo(False)
                
            else:
                escolha_jogador = dicionario_melhor_visualizacao[escolha_jogador]
                print(f"Você escolheu: {escolha_jogador}")            

                escolha_computador = random.choice(list(dicionario_melhor_visualizacao.values()))  
                print(f"O computador escolheu: {escolha_computador}")
                print("\n")
                return escolha_jogador, escolha_computador  
            

            
            
def determinar_vencedor(escolha_jogador, escolha_computador):    
        
    if escolha_jogador == escolha_computador:
        return "Empate"
    elif (escolha_jogador == 'Pedra' and escolha_computador == 'Tesoura') or \
         (escolha_jogador == 'Papel' and escolha_computador == 'Pedra') or \
         (escolha_jogador == 'Tesoura' and escolha_computador == 'Papel'):
        return "Você ganhou!"
    else:
        return "Você perdeu!"
    
    

def placar(resultado, pontos_jogador, pontos_computador, empate):
    
    if resultado == 'Você ganhou!':
        pontos_jogador += 1
    elif resultado == "Você perdeu!":
        pontos_computador += 1
    elif resultado == "Empate":
            empate += 1
    else:
        pass
            
    return pontos_jogador, pontos_computador, empate
                


def conferir_se_empate(resultado, contador):
    if resultado == 'Você ganhou!':
        contador += 1
    elif resultado == "Você perdeu!":
        contador += 1
    elif resultado == "Empate":
            contador = contador
    return contador
    

    
def estatisticas(pontos_jogador, pontos_computador, empate):
    
    total = pontos_jogador + pontos_computador + empate
    
    porcentagem_vitoria = f'{((pontos_jogador / total) * 100):.2f}%'
    porcentagem_derrota = f'{((pontos_computador / total) * 100):.2f}%'
    porcentagem_empate = f'{((empate / total) * 100):.2f}%'
    
    print('Estatísticas')
    
    print(f'Total de Partidas: {total}')
    print(f'Vitórias: {pontos_jogador} | {porcentagem_vitoria}')
    print(f'Derrotas: {pontos_computador} | {porcentagem_derrota}')
    print(f'Empates: {empate} | {porcentagem_empate}')
    print("\n")
    
    

def main():
    
    while True:
        menu_principal()
        escolha = input('Escolha uma opção: ')
        

        if escolha == '1':
            print("\n")
            print("Você escolheu Jogo Único.")
            print("\n")
            menu_jogo(False)
            # x puxa o primeiro return da função jogadas (escolha_jogador), enquanto y fica com o segundo (escolha_computador)
            x, y = jogadas(False)
            resultado = determinar_vencedor(x, y)
            print(resultado)
            print("\n")
            a, b, c = placar(resultado, 0, 0, 0)
            estatisticas(a, b, c)
            
            
        elif escolha == '2':
            print("\n")
            print("Você escolheu Vários Jogos.")
            print("\n")
            a, b, c = 0, 0, 0
            while True:
                menu_jogo(True)
                x, y = jogadas(True)
                # mesma logica do de cima: x puxa o primeiro return, enquanto y fica com o segundo
                if x == '4':
                    estatisticas(a, b, c)
                    break
                resultado = determinar_vencedor(x,y)
                print(resultado)
                print("\n")
                a, b, c = placar(resultado, a, b, c)
                estatisticas(a, b, c)


        elif escolha == '3':
            print("\n")
            print("Você escolheu Melhor de 3.")
            print("\n")
            a, b, c = 0, 0, 0
            contador = 0
            while contador < 3:
                menu_jogo(True)
                x, y = jogadas(True)
                if x == '4':
                    break
                resultado = determinar_vencedor(x,y)
                print(resultado)
                print("\n")
                contador = conferir_se_empate(resultado, contador)
                a, b, c = placar(resultado, a, b, c)
                estatisticas(a, b, c)
                
                if a == 2:
                    print("Parabéns, você ganhou a melhor de 3!")
                    print("\n")
                    estatisticas(a, b, c)
                    break
                elif b == 2:
                    print("O computador ganhou a melhor de 3!")
                    print("\n")
                    estatisticas(a, b, c)
                    break                


        elif escolha == '4':
            print("\n")
            print("Você escolheu Melhor de 5.")
            print("\n")
            a, b, c = 0, 0, 0
            contador = 0
            while contador < 5:
                menu_jogo(True)
                x, y = jogadas(True)
                if x == '4':
                    break
                resultado = determinar_vencedor(x,y)
                print(resultado)
                print("\n")
                contador = conferir_se_empate(resultado, contador)
                a, b, c = placar(resultado, a, b, c)
                estatisticas(a, b, c)
                
                
                if a == 3:
                    print("Parabéns, você ganhou a melhor de 5!")
                    print("\n")
                    estatisticas(a, b, c)
                    break
                elif b == 3:
                    print("O computador ganhou a melhor de 5!")
                    print("\n")
                    estatisticas(a, b, c)
                    break 


        elif escolha == '5':
            print("\n")
            print("Você escolheu Melhor de 7.")
            print("\n")
            a, b, c = 0, 0, 0
            contador = 0
            while contador < 7:
                menu_jogo(True)
                x, y = jogadas(True)
                if x == '4':
                    break
                resultado = determinar_vencedor(x,y)
                print(resultado)
                print("\n")
                contador = conferir_se_empate(resultado, contador)
                a, b, c = placar(resultado, a, b, c)
                estatisticas(a, b, c)
                
                if a == 4:
                    print("Parabéns, você ganhou a melhor de 7!")
                    print("\n")
                    estatisticas(a, b, c)
                    break
                elif b == 4:
                    print("O computador ganhou a melhor de 7!")
                    print("\n")
                    estatisticas(a, b, c)
                    break  

        elif escolha == '6':
            print("Saindo do jogo. Até a próxima!")
            break

        else:
            print("Escolha inválida. Tente novamente.")

        print("\n")

main()
```
